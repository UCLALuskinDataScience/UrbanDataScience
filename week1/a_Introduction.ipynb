## 3. Set up the Anaconda environment

Anaconda makes it easy to install Python (and R and other data science packages) in a virtual environment. That means that it won't interfere with your system installation of Python. It's easy to load in additional packages. And it works on Windows, Mac, and Linux.

<div class="alert alert-block alert-info">
If you are unable to install Anaconda or are uncomfortable doing so, please talk to me about alternatives. Jupyter Hub is a good option that runs in the cloud or on a UCLA server, but it is more constrained, especially with memory. You are likely to have problems with some of the more complex assignments.
</div>

You should already have [downloaded and installed Anaconda](https://www.anaconda.com/products/individual). 

First, [add conda forge to your Anaconda environment](https://conda-forge.org/docs/user/introduction.html#display-conda-forge-packages-in-anaconda-navigator). This gives you access to specialized packages such as `folium` (for mapping).

Second, [install additional packages](https://docs.anaconda.com/anaconda/navigator/tutorials/manage-packages/). For starters, we'll need the following: `cenpy`, `contextily`, `folium`, `geopandas`, `requests`, and `shapely`.

Third, go back to the Anaconda home screen. Find the icon for Jupyter Notebook. Click on Install (the first time), and then Launch.

## 4. Set up git
So how do you access the iPython notebooks that I create for lecture? One option is to simply download them manually from the course site. But that will be a pain. They often won't be ready until right before class.

The solution: git! 

Git is an incredibly powerful tool for version control and syncing. You can use it from the command line, but if you are just getting started, I recommend GitHub Desktop.

<div class="alert alert-block alert-info">
If you are unable to install GitHub Desktop or are uncomfortable using GitHub's website or software, please talk to me about alternatives.
</div>


1. You should already have a GitHub account. If not, [create one](https://github.com/join).
2. You should have access to the course repository. If not, I will add you.
3. Download [GitHub Desktop](https://desktop.github.com).
4. When you open it up, you should see the course repository (`UCLALuskinDataScience/UrbanDataScience`) in the list. Select it, and click on Clone. Choose the name of a new folder where you want to keep your course git folder.
5. By magic, you should see the course files appear. And whenever you want to fetch the latest version (i.e., before class), then click on Fetch origin.
6. Go to the Jupyter Notebook browser window that you opened before. Now, you should be able to open this file `/week1/a_Introduction.ipynb` and follow along.

There will be more on git soon, as you'll use it to submit an assignment.
